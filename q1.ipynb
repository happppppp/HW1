# Answer 1

In this question, we are asked to write a fucntion to compute Riemann sum. To solve this problem, we choose to calculate to left riemann sum. At first, we import the necessary libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Then we define a function f to calculate to given function from the user. In this example, we will calculate x*x function between 0 and 1 with 10 partitions.

```python
def f(value, function):
    x = value
    return eval(function)
```

Now, we are ready to write our function which will calculate the left riemann sum. To do so, we need to use upper left corner of our rectangles. Therefore, we will create N+1 intervals and omit the last one with the commands,

* x_left = x[:-1]
* y_left = y[:-1]

```python



def g(a,b,fn,N):
    x = np.linspace(a,b,N+1)
    y = f(x,fn)

    X = np.linspace(a,b,N+1)
    Y = f(X,fn)

    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.plot(X,Y,'b')
    x_left = x[:-1] # Left endpoints
    y_left = y[:-1]
    plt.plot(x_left,y_left,'b.',markersize=10)
    plt.bar(x_left,y_left,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
    plt.title('Left Riemann Sum, N = {}'.format(N))



    plt.show()

fn = input('function') 
a = int(input('interval left: '))
b = int(input('interval right: '))
N = int(input('interval partition: '))

g(a,b,fn,N)

 # Use n*N+1 points to plot the function smoothly
#func_text = input('Enter a function: ')
#func_eval = eval(func_text)
#f = lambda x: func_eval

```

We've shown the area and now we are ready to calculate it numerically:

```python
dx = (b-a)/N
x_left = np.linspace(a,b-dx,N)


print("Partition with",N,"subintervals.")
left_riemann_sum = np.sum(f(x_left,fn) * dx)
print("Left Riemann Sum:",left_riemann_sum)

```

# Answer 3

To answer this question, we will do something similar in the second answer. At first, we take a random number $a$ between $(-1,1)$ and calculate $b$ just as we did above. After that we check whether $1-a^2-b^2$ is equal to $0$ or not. We are doing this because **math.sqrt** does not take $0$ or negative values. If this value other than zero, we calculate $c$ as we calculate $b$. Finally we can print each value and show the sum is equal to $1$

```python
import matplotlib.pyplot as plt
import math, random, numpy as np
from mpl_toolkits.mplot3d import Axes3D

i=0
fig = plt.figure(figsize=(4,4))

ax = fig.add_subplot(111, projection='3d')


def f(a,b,c):
    e=np.array([])
    g=np.array([])
    i=np.array([])
    e=np.append(e,a)
    g=np.append(g,b)
    i=np.append(i,c)
    h=np.concatenate((e, g, i), axis=0)
    plt.xlim(-1.25,1.25)
    plt.ylim(-1.25,1.25)
    #ax.add_artist(circle1)
    ax.scatter(e,g,i)
    plt.grid(linestyle='--',)
    #plt.plot(e, g ,'b+')
    print(e*e+g*g+i*i)
    return h

N=int(input('How many n-tuples:'))   
for i in range(N):
    a = random.uniform(-1,1)
    b = random.uniform(-math.sqrt(1-(a*a)),math.sqrt(1-(a*a)))
    if (1-(a*a)-(b*b)) == 0:
        c=0
    else:
        c = math.sqrt(1-(a*a)-(b*b))
        
    print(f(a,b,c))
```
